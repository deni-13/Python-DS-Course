## Gorsellestirme

```python
import numpy as np
import matplotlib.pyplot as pyt
```

```python
np1=np.array([1,2,3,4])
```

```python
new=plt.figure(dpi=200)
newE=new.add_axes([0.1,0.1,0.5,0.9])
newE.plot(np1**3,label="np")
newE.plot(np1**4,label="np2")
newE.legend()
```

```python
#legend icine loc alıyor bunu deneyebiliriz
```

```python
#dpi---> cozunurluk plt.figure'un icine giriyor 
```

## Gorsellik İleri

```python
npp=np.linspace(1,10,20)
```

```python
npp
```

```python
npp2=npp**2
```

```python
npp2
```

```python
a,b=pyt.subplots()
b.plot(npp,npp2,color="#3A95A8")
b.plot(npp2,npp,"m")
```

```python
#rgb hex renk kodları

#color =string formatı hex vermek

```

```python
#CİZGİ KALINLIGI
a,b=plt.subplots()
b.plot(npp+5,linewidth=8)
b.plot(npp2+99,linewidth=8,linestyle=":")
b.plot(npp2+99,linewidth=16,linestyle=":",color="#000000",markersize=4)
```

```python
#marker---> + : gibi 
#marker size---> boyut
```
